# Stem Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,5,3,8,5]

plt.ylabel("Popularity",fontsize=14,color="orange")
plt.xlabel("Languages",fontsize=14,color="orange")
plt.title("Stem Plot",fontsize=14,color="blue")

plt.stem(x,y,linefmt="-.",markerfmt="ro",bottom=2,basefmt="orange",label="product")
plt.xticks(x,labels=["Python","Java","C","C++","JavaScript"])
plt.legend(loc=9,facecolor="yellow",edgecolor="pink",fontsize=10,framealpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[1,2,3,4,5]
y1=[1,5,3,8,5]

plt.ylabel("Popularity",fontsize=14,color="orange")
plt.xlabel("Languages",fontsize=14,color="orange")
plt.title("Stem Plot",fontsize=14,color="blue")

plt.stem(x1,y1,linefmt=":",markerfmt="r+",orientation="horizontal")
plt.yticks(x1,labels=["Python","C","C++","Java","JavaScript"])
# plt.xlim(0,10)
plt.axis([0,10,0,7])
plt.show()
```
