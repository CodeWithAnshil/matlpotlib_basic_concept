# Pie PLot

```python
import matplotlib.pyplot as plt
x=[20,30,10,40]
y=['python','C','C++','Java']
ex=[0.1,0.0,0.0,0.0]

plt.title("Pie Chart",fontsize=15)

plt.pie(x,labels=y,explode=ex,shadow=False,autopct="%0.f%%",textprops={"fontsize":15},labeldistance=1.2,rotatelabels=True)
plt.legend(loc=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,10,40]
z=[12]
y=['python','C','C++','Java']
ex=[0.1,0.0,0.0,0.0]

plt.title("Pie Chart",fontsize=15)

plt.pie(x,labels=y,explode=ex,shadow=False,autopct="%0.f%%",textprops={"fontsize":15},labeldistance=1.2,rotatelabels=True,radius=1)
plt.pie(z,colors="w",radius=0.4)
plt.legend(loc=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[20,30,10,40]
y=['python','C','C++','Java']
ex=[0.1,0.0,0.0,0.0]

plt.title("Pie Chart",fontsize=15)

plt.pie(x,labels=y,explode=ex,shadow=False,autopct="%0.f%%",textprops={"fontsize":15},labeldistance=1.2,rotatelabels=True)
cr=plt.Circle(xy=(0,0),radius=.4,facecolor="w")
plt.gca().add_artist(cr)
plt.legend(loc=1)
plt.show()
```
