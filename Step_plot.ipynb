# Step_Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,4,5,5,6]
y=[9,5,6,7,4,3]

plt.xlabel("Share Down",fontsize=15,color="blue",style="italic")
plt.ylabel("Share Up",fontsize=15,color="blue",style="italic")
plt.title("Share Market",fontsize=15,color="red")

plt.step(x,y,color="blue",marker="P",mfc="r",ms=12,label="Check")
plt.legend(loc=9,facecolor="pink",edgecolor="black")
plt.grid()
plt.show()
```
