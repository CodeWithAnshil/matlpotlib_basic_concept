```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,3,4,5,6,7])
y=np.array([3,3,4,6,3,8])

plt.fill_between(x,y)
# plt.plot(x,y)
plt.show()
```

```python
x1=np.linspace(0,2*np.pi,360)

y1=np.sin(x1)
y2=np.cos(x1)

plt.fill_between(x1,y1)
# plt.fill_between(x1,y2)
plt.plot(x1,y1,label="Sin(x)")

plt.legend(loc=9)
plt.show()
```
