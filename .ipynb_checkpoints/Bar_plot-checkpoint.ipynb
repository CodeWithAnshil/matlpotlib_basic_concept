# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["Python","C","C++","Java"]
y=[86,70,60,90]
z=[89,90,67,28]
width=0.5
p=np.arange(len(x))
p1=[i+width for i in p]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("AnshilTech")
plt.bar(p,z,width,color="g",edgecolor="b",alpha=0.7)
plt.bar(p1,y,width,color="r",edgecolor="b",alpha=0.7)
plt.show()
```
