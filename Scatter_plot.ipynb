# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
floor1=[20,30,20,54,34,6,89]
floor2=[30,89,49,20,89,59,29]
floor3=[40,39,49,24,69,79,49]
color=[12,34,45,24,33,53,34]
sizes=[122,342,166,233,267,146,175]

plt.xlabel("Days",fontsize=15)
plt.ylabel("Pepole",fontsize=15)
plt.title("Scatter Bar",fontsize=25)
plt.scatter(day,floor1,c=color,s=sizes,cmap="viridis",label="floor1")
plt.scatter(day,floor2,s=sizes,color='pink',label="floor2")
plt.scatter(day,floor3,s=sizes,color='orange',label="floor2")
t=plt.colorbar()
t.set_label("ColorLabel",fontsize=15)
plt.legend()
plt.show()
```
