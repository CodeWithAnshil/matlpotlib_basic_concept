# Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
area1=[2,8,7,5]
area2=[8,9,8,9]
y=[2,3,7,6]
plt.stackplot(x,y,area1,area2,labels=["Area1","Area2","Area3"],colors=["Blue","red","green"],edgecolor="black")
plt.legend(loc=2,facecolor="pink",edgecolor="black")
plt.show()
```
