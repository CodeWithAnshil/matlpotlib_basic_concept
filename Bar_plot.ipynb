# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["Python","C","C++","Java"]
y=[86,70,60,90]
z=[89,90,67,28]
width=0.2
p=np.arange(len(x))
p1=[i+width for i in p]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("Bar Plot")
plt.bar(p,z,width,color="g",edgecolor="b",alpha=0.7,label="Past")
plt.bar(p1,y,width,color="r",edgecolor="b",alpha=0.7,label="present")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
x1=["Python","C","C++","Java"]
y1=[86,70,60,50]
z1=[89,90,67,70]
width=0.2
p=np.arange(len(x1))
p1=[i+width for i in p]
plt.xlabel("Language",fontsize=15)
plt.ylabel("Percentage",fontsize=15)
plt.title("AnshilTech")
plt.barh(p,z1,width,color="g",edgecolor="b",alpha=0.7,label="Past")
plt.barh(p1,y1,width,color="r",edgecolor="b",alpha=0.7,label="present")

plt.yticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
